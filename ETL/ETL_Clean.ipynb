2. TRANSFORM (TRATAMIENTO DE DATOS)

- Diccionarios y Listas anidadas

```python
# Función para inspeccionar diccionarios y listas en las columnas
def inspeccionar_columnas(df, n=5):
    columnas_diccionarios = []
    columnas_listas = []
    
    for columna in df.columns:
        for i in range(n):
            valor = df[columna].iloc[i]
            if pd.notnull(valor) and isinstance(valor, str):
                try:
                    eval_valor = ast.literal_eval(valor)
                    if isinstance(eval_valor, dict):
                        columnas_diccionarios.append(columna)
                        break
                    elif isinstance(eval_valor, list):
                        columnas_listas.append(columna)
                        break
                except (ValueError, SyntaxError):
                    continue
    
    return columnas_diccionarios, columnas_listas
```

```python
# Inspeccionar diccionarios y listas en el DataFrame de películas
columnas_diccionarios, columnas_listas = inspeccionar_columnas(movies_df)

print("Columnas que contienen diccionarios:")
print(columnas_diccionarios)

print("\nColumnas que contienen listas:")
print(columnas_listas)
```

```python
# Inspeccionar diccionarios y listas en el DataFrame de créditos
columnas_diccionarios, columnas_listas = inspeccionar_columnas(credits_df)

print("Columnas que contienen diccionarios:")
print(columnas_diccionarios)

print("\nColumnas que contienen listas:")
print(columnas_listas)
```

```python
# Función para desanidar diccionarios y listas en una columna
def desanidar_columna(df, columna):
    rows = []  # Inicializar una lista para almacenar las filas desanidadas
    for idx, line in enumerate(df[columna]):
        if isinstance(line, str) and line.startswith(('[' ,'{')):  # Verifica si es una cadena que puede ser evaluada como lista o diccionario
            try:
                parsed_list = ast.literal_eval(line)  # Evalúa la cadena como una lista o un diccionario de Python
                if isinstance(parsed_list, list):  # Si es una lista
                    for item in parsed_list:
                        row = df.loc[idx].to_dict()  # Convierte la fila actual del DataFrame a un diccionario
                        if isinstance(item, dict):
                            for k, v in item.items():
                                row[f"{columna}_{k}"] = v  # Añade cada clave del diccionario como nueva columna con prefijo
                        else:
                            row[f"{columna}_item"] = item  # Añade el valor como una nueva columna si no es un diccionario
                        rows.append(row)  # Añade la fila desanidada a la lista de filas
                elif isinstance(parsed_list, dict):  # Si es un diccionario
                    row = df.loc[idx].to_dict()  # Convierte la fila actual del DataFrame a un diccionario
                    for k, v in parsed_list.items():
                        row[f"{columna}_{k}"] = v  # Añade cada clave del diccionario como nueva columna con prefijo
                    rows.append(row)  # Añade la fila desanidada a la lista de filas
            except (ValueError, SyntaxError):
                rows.append(df.loc[idx].to_dict())  # Si hay un error en la evaluación, añade la fila original
        else:
            rows.append(df.loc[idx].to_dict())  # Si la cadena no es una lista o diccionario, añade la fila original
    expanded_df = pd.DataFrame(rows)  # Crea un nuevo DataFrame con las filas desanidadas
    return expanded_df  # Devuelve el DataFrame expandido
```

```python
# Desanidar la columna 'belongs_to_collection' en el DataFrame de películas
movies_df = desanidar_columna(movies_df, 'belongs_to_collection')
```

```python
# Desanidar la columna 'genres' en el DataFrame de películas
movies_df = desanidar_columna(movies_df, 'genres')
```

```python
# Desanidar la columna 'cast' en el DataFrame de créditos
credits_cast_df = desanidar_columna(credits_df, 'cast')
```

```python
# Desanidar la columna 'crew' en el DataFrame de créditos
credits_crew_df = desanidar_columna(credits_df, 'crew')
```

- Valores nulos de los campos revenue, budget rellenados con número 0.

```python
# Convertir 'revenue' y 'budget' a valores numéricos (float), manejando errores
movies_df['revenue'] = pd.to_numeric(movies_df['revenue'], errors='coerce').fillna(0)
movies_df['budget'] = pd.to_numeric(movies_df['budget'], errors='coerce').fillna(0)
```

- Eliminación de Valores nulos del campo release date 

```python
# Eliminar las filas con valores nulos en la columna 'release_date'
movies_df = movies_df.dropna(subset=['release_date']).copy()
```

- Fechas con formato AAAA-mm-dd, creación de columna release_year

```python
# Convertir 'release_date' al formato AAAA-mm-dd y verificar el tipo de dato
movies_df['release_date'] = pd.to_datetime(movies_df['release_date'], format='%Y-%m-%d', errors='coerce')
```

```python
# Asegúrate de que 'release_date' es un datetime
print(movies_df['release_date'].dtype)  # Esto debería imprimir 'datetime64[ns]'
```

```python
# Eliminar las filas con fechas inválidas que no pudieron convertirse
movies_df = movies_df.dropna(subset=['release_date']).copy()
```

```python
# Crear la columna 'release_year' extrayendo el año de 'release_date'
movies_df['release_year'] = movies_df['release_date'].dt.year
```

- Creación de columna retorno de inversión, llamada return = revenue / budget, cuando no hay datos disponibles para calcularlo, toma el valor 0.

```python
# Crear la columna 'return' calculando revenue / budget
movies_df['return'] = (movies_df['revenue'] / movies_df['budget']).replace([float('inf'), float('-inf')], 0).fillna(0)
```
